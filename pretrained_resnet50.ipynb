```python
#MODEL

input_shape=(100,100,3)

model = ResNet50( include_top=False , input_shape= input_shape , classes=4)
```

```python
#Flatten output layer of Resnet
flattened = tf.keras.layers.Flatten()(model.output)

#Fully connected layer 1
fc1 = tf.keras.layers.Dense(128, activation='relu', name="AddedDense1")(flattened)

#Fully connected layer, output layer
fc2 = tf.keras.layers.Dense(4, activation='softmax', name="AddedDense2")(fc1)

model = tf.keras.models.Model(inputs=model.input, outputs=fc2)
```

```python
"""train_y= ((df1[['category','midcategory','maincategory','color']]).iloc[100:1000]).astype('str')
train_x= ((df1[['productid']]).iloc[100:1000]).astype('str')

test_x= (df1[['category','midcategory','maincategory','color']]).iloc[0:100]
test_y= (df1[['productid']]).iloc[0:100]
"""
def append_ext(fn):
    return str(fn)+".png"


train= df1.iloc[0:10]
train["productid"]=train["productid"].apply(append_ext)
train = train.astype('str')
train.replace(np.nan, 0)
```

```python
datagen = ImageDataGenerator(rescale=1./255)

train_generator=datagen.flow_from_dataframe(train, 
                                            directory='/content/gdrive/MyDrive/Modanisa/dataset/',
                                            x_col='productid',
                                            y_col=['category','midcategory','maincategory','color'],
                                            class_mode="multi_output",
                                            color_mode='rgb',
                                            target_size=(100,100),
                                            batch_size=10)
```

```python
model.compile(optimizer="rmsprop", loss="categorical_crossentropy",metrics=['accuracy'])
```

```python
model.fit(train_generator ,epochs=10)
```
